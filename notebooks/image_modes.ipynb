```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import tensorflow as tf
import matplotlib.pyplot as plt
import PIL
import numpy as np
```

## PIL

```python
bw_path = r"..\data\ILSVRC2012_img_train\n01491361\n01491361_3490.JPEG"
bw = PIL.Image.open(bw_path)
print(f"Image mode: {bw.mode}")
print(np.asarray(bw).shape)
bw
```

```python
shark = tf.io.read_file(bw_path)
shark = tf.image.decode_jpeg(shark, channels=3)
print(shark.shape)
plt.imshow(shark)
```

```python
np.array_equal(np.asarray(bw.convert("RGB")), shark.numpy())
```

```python
(shark.numpy().astype(np.float) - np.asarray(bw.convert("RGB")).astype(np.float))
```
