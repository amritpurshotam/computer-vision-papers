# PCA Colour Augmentation Demo (Tensorflow 2)

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

from src.utilities.image import fancy_pca
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def plot_pca_examples(image_path: str, num_imgs: int, alpha_std: float):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)

    plt.figure(figsize=(40,8))
    ax = plt.subplot(1, num_imgs, 1)
    ax.axis("off")
    plt.imshow(image)

    for i in range(1, num_imgs):
        pca = fancy_pca(image, False, 0.3)
        ax = plt.subplot(1, num_imgs, i+1)
        ax.axis("off")
        plt.imshow(pca)
```

```python
plot_pca_examples(image_path="bird.jpg", num_imgs=5, alpha_std=0.3)
```

```python
plot_pca_examples("chair.png", num_imgs=6, alpha_std=3)
```
